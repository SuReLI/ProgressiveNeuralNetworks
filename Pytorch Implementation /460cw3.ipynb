### Define model

```python
# *CODE FOR PART 1.1 IN THIS CELL*

### Choose a value for the latent space dimension and use it in your model
hidden_size = None #we don't use this variable but the latent space we choose for one (3,32,32) image is (128,2,2)
###


        
      
class CAE(nn.Module):
    def __init__(self):
        super(CAE, self).__init__()
        
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################
        self.encoder = nn.Sequential(
        nn.Conv2d(3,64,3,stride = 1, padding = 1), 
        nn.ReLU(), 
        nn.BatchNorm2d(64),
        nn.MaxPool2d(2,stride = 2), 
        nn.Conv2d(64,128,3,stride = 1, padding = 1), 
        nn.ReLU(), 
        nn.BatchNorm2d(128),
        nn.MaxPool2d(2,stride = 2),
        nn.Conv2d(128,256,3,stride = 1, padding = 1), 
        nn.ReLU(), 
        nn.BatchNorm2d(256),
        nn.MaxPool2d(2,stride = 2),
        nn.Conv2d(256,128,3,stride = 1, padding = 1), 
        nn.ReLU(), 
        nn.BatchNorm2d(128),
        nn.MaxPool2d(2,stride = 2))
        
        self.decoder = nn.Sequential(
        nn.ConvTranspose2d(128,128,4,stride = 2, padding = 0),
        nn.ReLU(), 
        nn.BatchNorm2d(128), 
        nn.ConvTranspose2d(128,64,4,stride = 2, padding = 0), 
        nn.ReLU(), 
        nn.BatchNorm2d(64),
        nn.ConvTranspose2d(64,32,3,stride = 2, padding = 0), 
        nn.ReLU(), 
        nn.BatchNorm2d(32),
        nn.ConvTranspose2d(32,3,4,stride = 1, padding = 0), 
        nn.Tanh())
        
        #######################################################################
        #                       ** END OF YOUR CODE **
        #######################################################################       

    def encode(self, x):
        
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################
        z = self.encoder(x)
        return z
        #######################################################################
        #                       ** END OF YOUR CODE **
        #######################################################################   
    
    def decode(self, z):
        
        #######################################################################
        #                       ** START OF YOUR CODE **
        #######################################################################
        x_recon = self.decoder(z)
        return x_recon
        #######################################################################
        #                       ** END OF YOUR CODE **
        #######################################################################      

    def forward(self, x):
        z = self.encode(x)
        recon = self.decode(z)
        return recon
```

### Define Loss function

```python
criterion = nn.MSELoss(reduction='mean')  # can we use any other loss here? You are free to choose.
def loss_function_CAE(recon_x, x):
    recon_loss = criterion(recon_x, x)
    return recon_loss
```

```python
train_losses = []
model.train()
for epoch in range(num_epochs):
    train_loss = 0
    for batch_idx, data in enumerate(loader_train):
        img, _ = data
        img = img.to(device)
        optimizer.zero_grad()
        # forward
        recon_batch = model(img)
        loss = loss_function_CAE(recon_batch, img)
        # backward
        loss.backward()
        train_loss += loss.item()
        optimizer.step()
    # print out losses and save reconstructions for every epoch
    print('epoch [{}/{}], loss:{:.4f}'
          .format(epoch + 1, num_epochs, train_loss / len(loader_train)))
    recon = model(fixed_input.to(device))
    recon = denorm(recon.cpu())
    save_image(recon, './CW/CAE/reconstructed_epoch_{}.png'.format(epoch))
    train_losses.append(train_loss/ len(loader_train))

# save the model and the loss values
np.save('./CW/CAE/train_losses.npy', np.array(train_losses))
torch.save(model.state_dict(), './CW/CAE/CAE_model.pth')
```
